```python
import pandas as pd
```

```python
df = pd.read_csv("olympic.csv")
df
```

```python
import matplotlib.pyplot as plt
```

```python
# Histogram
plt.hist(df["Medal"],histtype = "stepfilled")
plt.xlabel("Type of Medal")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar graph
plt.bar(df["Medal"],df["Year"],color = "c")
plt.xlabel("Type of Medal")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pie Chart 
plt.pie(df["Year"][:10],labels = df["Event"][:10], autopct = "% 1.1f%%")
plt.show()
```

```python
# Box plot
plt.boxplot(df["Year"])
plt.xlabel("Frequency")
plt.ylabel("Year")
plt.show()
```

```python
# Line Plot
plt.plot(df["Year"][:300])
plt.xlabel("Frequency")
plt.ylabel("Year")
plt.show()
```

```python
# Scatter Plot
plt.scatter(df["Year"][:20],df["Medal"][:20])
plt.show()
```
